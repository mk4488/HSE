### Задание 1

Дана переменная, в которой хранится слово из латинских букв. Напишите код, который
выводит на экран:
- среднюю букву, если число букв в слове нечётное;
- две средних буквы, если число букв чётное.

```python
'''функция возвращающая букву/буквы по заданию'''
def get_middle(word):
    length = len(word)
    middle = length // 2
    if length % 2 == 0:
        return word[middle - 1:middle + 1]
    else:
        return word[middle]
```

```python
'''проверка'''
word1 = 'test'
word2 = 'testing'

print(f'в слове {word1} количество букв = {len(word1)}. Результат: {get_middle(word1)}')
print(f'в слове {word2} количество букв = {len(word2)}. Результат: {get_middle(word2)}')
```

### Задание 2

Вы делаете MVP (минимально жизнеспособный продукт) dating-сервиса. У вас есть список юношей и девушек.
Выдвигаем гипотезу: лучшие рекомендации получатся, если просто отсортировать имена по алфавиту и познакомить людей с одинаковыми индексами после сортировки.
Но вы не будете никого знакомить, если кто-то может остаться без пары.

```python
'''функция для пар сервиса'''
def dating_pairs(boys, girls):
    if len(boys) != len(girls):
        return 'Внимание, кто-то может остаться без пары!'
    
    sorted_boys = sorted(boys)
    sorted_girls = sorted(girls)

    pairs = []
    for i in range(len(sorted_boys)):
        pairs.append(f"{sorted_boys[i]} и {sorted_girls[i]}")

        result = "Идеальные пары:\n" + "\n".join(pairs)
    return result
```

```python
'''проверка 1'''
boys = ['Peter', 'Alex', 'John', 'Arthur', 'Richard']
girls = ['Kate', 'Liza', 'Kira', 'Emma', 'Trisha']

print(dating_pairs(boys, girls))
```

```python
'''проверка 2'''
boys = ['Peter', 'Alex', 'John', 'Arthur', 'Richard', 'Michael']
girls = ['Kate', 'Liza', 'Kira', 'Emma', 'Trisha']

print(dating_pairs(boys, girls))
```
