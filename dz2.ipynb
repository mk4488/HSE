### Задание 1

Реализуйте функцию sum_distance(from, to), которая суммирует все числа от
значения from до величины to включительно.Примечание. Если пользователь задаст первое число, которое окажется
больше второго, просто поменяйте их местами.

```python
'''функция суммирующая все числа от значения from до величины to включительно'''
def sum_distance(start, end):
    if start > end:
        start, end = end, start

    total = 0
    for number in range(start, end + 1):
        total += number
    
    return total
```

```python
'''проверка 1'''
result1 = sum_distance(1, 5)
print(f'sum_distance(1, 5) = {result1}') 
```

```python
'''проверка 2'''
result2 = sum_distance(5, 1)
print(f'sum_distance(5, 1) = {result2}') 
```

### Задание 2

Напишите код на Python для решения следующей задачи.
Реализуйте функцию trim_and_repeat(), которая принимает три параметра:
- строку;
- offset — число символов, на которое нужно обрезать строку слева;
- repetitions — сколько раз нужно повторить строку перед возвратом
получившейся строки.

Число символов для среза по умолчанию равно 0, а количество повторений — 1.
Функция должна возвращать полученную строку.

```python
'''функция возврата строки'''
def trim_and_repeat(text, offset=0, repetitions=1):
    trim_text = text[offset:]
    result = trim_text * repetitions
    return result
```

```python
'''проверка 1'''
result1 = trim_and_repeat('test')
print(f'trim_and_repeat(test) = "{result1}"') 
```

```python
'''проверка 2. С обрезкой строки, без повторений'''
result2 = trim_and_repeat('test', offset=2)
print(f'trim_and_repeat(test, offset=2) = "{result2}"') 
```

```python
'''проверка 3. Без обрезки строки, с повторениями'''
result3 = trim_and_repeat('test', repetitions=3)
print(f'trim_and_repeat(test, repetitions=3) = "{result3}"')
```

```python
'''проверка 4. С обрезкойстроки, с повторениями'''
result4 = trim_and_repeat('test', offset=2, repetitions=3)
print(f'trim_and_repeat(test, offset=2, repetitions=3) = "{result4}"') 
```
