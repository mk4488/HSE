### Задание 1

Переведите содержимое файла purchase_log.txt в словарь purchases вида:
{'1840e0b9d4': 'Продукты', ...}

```python
import json
```

```python
with open('purchase_log.txt') as f:
    print([next(f) for x in range(5)])
```

```python
purchases = {}
i = 0
with open('purchase_log.txt') as f:
    for line in f:
        data = json.loads(line.strip())
        user_id = data.get('user_id')
        category = data.get('category')
        if user_id and category and user_id != 'user_id':
            purchases[user_id] = category

'''Добавил f-строку для точного соотвествия условиям в ДЗ:
purchases:
1840e0b9d4 ‘Продукты‘
4e4f90fcfb ‘Электроника‘'''
print('purchases:')
for user_id, category in list(purchases.items())[:2]:
    print(user_id, f"'{category}'")
```

### Задание 2

Для каждого user_id в файле visit_log.csv определите третий столбец с категорией покупки (если покупка была, сам файл visit_log.csv изменять не надо). Запишите в файл funnel.csv визиты из файла visit_log.csv, в которых были покупки с указанием категории.

```python
f = open('visit_log.csv', 'r')
content = f.readlines()
content[:5]
```

```python
with open('visit_log.csv', 'r', encoding='utf-8') as visit_f,\
     open('funnel.csv', 'w', encoding='utf-8') as funnel_f:
        funnel_f.write(visit_f.readline().strip() + ',category\n')
        for line in visit_f:
            line = line.strip()
            if line:
                user_id, source = line.split(',')
                if user_id in purchases: #проверка на наличие покупкив словаре покупок по user_id
                    funnel_f.write(f'{user_id},{source},{purchases[user_id]}\n')
```
